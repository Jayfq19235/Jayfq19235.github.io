```python
url_base= "https://api.stlouisfed.org/fred/series/observations?series_id={}&api_key=b3da81ef3fcd3b629670e247f9e29a75&file_type=json"
```

```python
file_base= "Fred-{}.json"
```

```python
# Choosing which series I would like data from
```

```python
series = ['FPCPITOTLZGUSA','GDP','UNRATE','TRESEGUSM052N']
```

```python
# Beginning loop
for i in series:


  # Setting urls and checking result
  URL= url_base.format(i)
  print(URL)

  # Requesting html
  data = requests.get(URL).json()
  print(data)

  # Setting filename

  fileName = file_base.format(i)
  print(fileName)

  # Opening file i.e selecting name we want to save under and then saving data to this file
  
  with open(fileName, 'w', encoding='utf-8') as f:
    json.dump(data,f, ensure_ascii=False, indent=4)

  # Batch downloading locally

  files.download(fileName)

```
